```python
# Cell 4 - Load a candidate CSV from Friend1 outputs (auto-detect) or create a small sample
import os
import pandas as pd

# Candidate filenames (from Friend1 notebook)
candidates = [
    "news_dataset.csv",
    "stock_features_with_llm.csv",
    "stock_features.csv",
    "stock_data_clean.csv",
    "stock_predictions_with_insights.csv",
    "sft_dataset.csv",
    "aapl_2022_updated.csv"
]

found_path = None
for fn in candidates:
    p = os.path.join(BASE_DIR, fn)
    if os.path.exists(p):
        found_path = p
        break

# if nothing from candidate list, use any CSV in BASE_DIR (prefer ones with 'stock' or 'news')
if found_path is None:
    csvs = [f for f in os.listdir(BASE_DIR) if f.lower().endswith(".csv")]
    # prefer names containing 'stock' or 'news'
    prioritized = [f for f in csvs if "stock" in f.lower()] + [f for f in csvs if "news" in f.lower()]
    if prioritized:
        found_path = os.path.join(BASE_DIR, prioritized[0])
    elif csvs:
        found_path = os.path.join(BASE_DIR, csvs[0])

if found_path:
    print("Loading CSV:", found_path)
    # load without forcing parse_dates to avoid parse errors; convert later if possible
    df = pd.read_csv(found_path)
    print("Columns:", list(df.columns)[:40])
    print("Preview:")
    display(df.head(5))
else:
    # No suitable CSV found -> create a synthetic news dataset (same as Friend2 earlier)
    print("No CSV found in BASE_DIR. Creating synthetic sample news dataset for demo.")
    dates = pd.date_range(end=pd.Timestamp.today(), periods=200)
    tickers = ['AAPL', 'MSFT', 'TSLA']
    rows = []
    for d in dates:
        for t in tickers:
            rows.append({'ticker': t, 'date': d.strftime('%Y-%m-%d'),
                         'text': f"{t} news on {d.date()}: company released product updates and reported earnings."})
    df = pd.DataFrame(rows)
    sample_path = os.path.join(BASE_DIR, "news_dataset.csv")
    df.to_csv(sample_path, index=False)
    print("Sample dataset written to", sample_path)

# Ensure we have 'ticker' and 'text' columns for DeepSeek parts; if not, try to synthesize a 'text' column
if 'ticker' not in df.columns:
    if 'Ticker' in df.columns:
        df.rename(columns={'Ticker': 'ticker'}, inplace=True)
    else:
        # try to infer ticker column by common names
        for col in df.columns:
            if col.lower() in ('symbol', 'tick', 'stock'):
                df.rename(columns={col: 'ticker'}, inplace=True)
                break
if 'text' not in df.columns:
    # attempt to synthesize text from available columns (headline, summary, content, or numeric columns)
    text_col = None
    for c in ['headline', 'content', 'summary', 'article', 'news']:
        if c in df.columns:
            text_col = c
            break
    if text_col:
        df['text'] = df[text_col].astype(str)
        print("Synthesized 'text' from column:", text_col)
    else:
        # if no textual columns, create simple descriptive text rows using numeric/price columns if present
        if 'close' in [c.lower() for c in df.columns] or 'close_price' in [c.lower() for c in df.columns] or 'Close Price' in df.columns:
            # build basic text from price columns
            def make_text(r):
                parts = []
                if 'ticker' in r: parts.append(str(r.get('ticker')))
                for col in ['Close Price', 'close', 'close_price']:
                    if col in r and pd.notnull(r[col]):
                        parts.append(f"price {r[col]}")
                return " ".join(parts)
            df['text'] = df.apply(make_text, axis=1)
            print("Synthesized 'text' from price column(s).")
        else:
            # final fallback: short placeholder text
            df['text'] = df.apply(lambda r: f"row {r.name} data", axis=1)
            print("Created placeholder 'text' column.")

# Convert date column if present
if 'date' in df.columns:
    try:
        df['date'] = pd.to_datetime(df['date'])
    except Exception:
        # leave it, will handle downstream
        pass

print("Final df columns (used by DeepSeek):", df.columns.tolist())
```

```python
# Cell 5 - Ensure date is datetime and sort by ticker/date (safe handling for multiple column-name variants)
import pandas as pd

# normalize column names: leave original but also work with variants
if 'date' in df.columns:
    df['date'] = pd.to_datetime(df['date'], errors='coerce')
else:
    # try common alternatives
    for c in df.columns:
        if 'date' in c.lower() or 'time' in c.lower():
            df.rename(columns={c: 'date'}, inplace=True)
            df['date'] = pd.to_datetime(df['date'], errors='coerce')
            break

# ensure ticker present
if 'ticker' not in df.columns:
    for c in df.columns:
        if c.lower() in ('symbol','ticker','tick','stock'):
            df.rename(columns={c: 'ticker'}, inplace=True)
            break

df = df.sort_values(['ticker', 'date'] if 'date' in df.columns else ['ticker']).reset_index(drop=True)
print("After sort - rows per ticker:")
if 'ticker' in df.columns:
    print(df.groupby('ticker').size().head())
else:
    print("No 'ticker' column found; rows:", len(df))

display(df.head(6))
```

```python
# Cell 6 - Load sentence-transformer and compute normalized embeddings
from sentence_transformers import SentenceTransformer
import numpy as np
import os

MODEL_NAME = 'all-MiniLM-L6-v2'  # small & fast: good for prototypes
print("Loading model:", MODEL_NAME)
model = SentenceTransformer(MODEL_NAME)

documents = df['text'].astype(str).tolist()
print("Encoding", len(documents), "documents (may take a minute)...")
embeddings = model.encode(documents, batch_size=64, show_progress_bar=True, convert_to_numpy=True)
embeddings = embeddings.astype('float32')

# L2-normalize - inner product in FAISS will behave like cosine
norms = np.linalg.norm(embeddings, axis=1, keepdims=True)
norms[norms == 0] = 1.0
embeddings = embeddings / norms

EMB_PATH = os.path.join(BASE_DIR, 'deepseek_embeddings.npy')
np.save(EMB_PATH, embeddings)
print("Saved embeddings to", EMB_PATH)
```

```python
 # Cell 7 - Build FAISS index and persist index + metadata into Friend1 folder
import faiss, joblib, os
EMB_DIM = embeddings.shape[1]
print("Embedding dim:", EMB_DIM)

index = faiss.IndexFlatIP(EMB_DIM)   # inner product on normalized vectors => cosine
index.add(embeddings)
print("Index contains", index.ntotal, "vectors")

INDEX_PATH = os.path.join(BASE_DIR, 'deepseek_index.faiss')
faiss.write_index(index, INDEX_PATH)
print("Saved FAISS index to", INDEX_PATH)

META_PATH = os.path.join(BASE_DIR, 'deepseek_metadata.pkl')
joblib.dump(df.reset_index(drop=True), META_PATH)
print("Saved metadata to", META_PATH)
```

```python
# Cell 8 - query_deepseek helper that reads index & metadata already loaded in memory
import numpy as np
import pandas as pd

def query_deepseek(query_text, top_k=5, return_dataframe=True):
    q_emb = model.encode(str(query_text), convert_to_numpy=True).astype('float32')
    q_emb = q_emb / (np.linalg.norm(q_emb) + 1e-9)
    D, I = index.search(np.array([q_emb]), top_k)
    results = []
    for score, idx in zip(D[0], I[0]):
        idx = int(idx)
        if idx < 0 or idx >= len(df):
            continue
        meta = df.iloc[idx].to_dict()
        meta['score'] = float(score)
        results.append(meta)
    if return_dataframe:
        return pd.DataFrame(results)
    return results

# Quick test query
print("Sample query: 'earnings beat' (top 5)")
display(query_deepseek("earnings beat", top_k=5))
```

```python
# Cell 11 - Build a compact news_signal per (ticker, date)
from collections import defaultdict
import numpy as np
mapping = defaultdict(list)

# Ensure df rows align with embeddings (they should)
for i, row in df.reset_index(drop=True).iterrows():
    # get date as date if available
    date_key = None
    if 'date' in row and not pd.isna(row['date']):
        try:
            date_key = pd.to_datetime(row['date']).date()
        except Exception:
            date_key = row['date']
    else:
        # fallback to index-based pseudo-date
        date_key = i
    key = (row.get('ticker', 'UNKNOWN'), date_key)
    mapping[key].append(embeddings[i])

# mean vectors
mean_emb = {}
for k, arr in mapping.items():
    arr = np.stack(arr)
    v = arr.mean(axis=0)
    v = v / (np.linalg.norm(v) + 1e-9)
    mean_emb[k] = v

# prototype prompts for "positive" and "negative" sentiment-like signals
pos_proto = model.encode('positive earnings good growth beat', convert_to_numpy=True).astype('float32')
neg_proto = model.encode('negative loss declined miss', convert_to_numpy=True).astype('float32')
pos_proto /= np.linalg.norm(pos_proto)
neg_proto /= np.linalg.norm(neg_proto)

news_rows = []
for (t, d), v in mean_emb.items():
    score = float(np.dot(v, pos_proto) - np.dot(v, neg_proto))
    news_rows.append({'ticker': t, 'date': pd.to_datetime(d) if not isinstance(d, int) else d, 'news_signal': score})

news_df = pd.DataFrame(news_rows)
print("news_df rows:", len(news_df))
display(news_df.head(6))
```

```python
# Cell 12 - Load prices (prefer Friend1's stock_data_clean.csv or stock_features.csv) then merge news_signal
import os, pandas as pd, numpy as np

# Candidate price files produced by Friend1
price_candidates = [
    os.path.join(BASE_DIR, 'stock_data_clean.csv'),
    os.path.join(BASE_DIR, 'stock_features.csv'),
    os.path.join(BASE_DIR, 'stock_features_with_llm.csv'),
    os.path.join(BASE_DIR, 'stock_predictions_with_insights.csv'),
    os.path.join(BASE_DIR, 'prices.csv')
]

price_path = None
for pc in price_candidates:
    if os.path.exists(pc):
        price_path = pc
        break

if price_path:
    print("Loading prices from", price_path)
    price_df = pd.read_csv(price_path)
else:
    # fallback: if df has price columns, try to use it; else create synthetic
    available_price_cols = [c for c in df.columns if 'close' in c.lower() or 'close price' in c.lower() or 'close_price' in c.lower()]
    if available_price_cols:
        print("Using columns from earlier DF as price source:", available_price_cols)
        price_df = df.copy()
        # ensure 'close' column exists
        if 'Close Price' in price_df.columns and 'close' not in price_df.columns:
            price_df['close'] = price_df['Close Price']
        elif 'close' not in price_df.columns:
            # pick the first close-like column
            price_df['close'] = price_df[available_price_cols[0]]
    else:
        print("No price file found; creating synthetic random-walk prices per ticker.")
        price_parts = []
        for t in df['ticker'].unique():
            sub = df[df['ticker'] == t].copy()
            dates = pd.to_datetime(sub['date'].drop_duplicates().sort_values()) if 'date' in sub.columns else pd.date_range(end=pd.Timestamp.today(), periods=200)
            n = len(dates)
            rng = np.random.RandomState(abs(hash(t)) % 2**32)
            returns = rng.normal(loc=0.0002, scale=0.02, size=n)
            price = 100 * np.cumprod(1 + returns)
            price_parts.append(pd.DataFrame({'ticker': t, 'date': dates.values, 'close': price}))
        price_df = pd.concat(price_parts).reset_index(drop=True)
        save_path = os.path.join(BASE_DIR, 'prices.csv')
        price_df.to_csv(save_path, index=False)
        print("Synthetic prices saved to", save_path)

# Normalize column names: lower + replace spaces
price_df.columns = [c.strip() for c in price_df.columns]
# find date column
if 'date' not in price_df.columns:
    for c in price_df.columns:
        if 'date' in c.lower():
            price_df.rename(columns={c: 'date'}, inplace=True)
            break
# find close column
if 'close' not in price_df.columns:
    for c in price_df.columns:
        if 'close' in c.lower():
            price_df.rename(columns={c: 'close'}, inplace=True)
            break

# Ensure date is datetime.date for merge
price_df['date'] = pd.to_datetime(price_df['date'], errors='coerce').dt.date

# Prepare news_df date column similarly
if 'date' in news_df.columns:
    news_df['date'] = pd.to_datetime(news_df['date'], errors='coerce').dt.date

# Merge news_signal into price_df
price_df = price_df.merge(news_df, how='left', on=['ticker', 'date'])
price_df['news_signal'] = price_df['news_signal'].fillna(0.0)

print("price_df rows:", len(price_df))
display(price_df.head(6))
```

```python
# Cell 14 - TradingEnv: simple Gym env that consumes price + news_signal
import gym
from gym import spaces
import numpy as np
import pandas as pd

class TradingEnv(gym.Env):
    metadata = {'render.modes': ['human']}

    def __init__(self, df, ticker, initial_balance=10000):
        super().__init__()
        self.ticker = ticker
        self.df = df[df['ticker'] == ticker].sort_values('date').reset_index(drop=True)
        if len(self.df) < 30:
            raise ValueError('Need at least 30 price days for the environment; got %d' % len(self.df))
        self.current_step = 1
        self.action_space = spaces.Discrete(3)  # 0:flat, 1:long, 2:short
        self.observation_space = spaces.Box(low=-np.inf, high=np.inf, shape=(4,), dtype=np.float32)

    def _get_obs(self):
        row = self.df.loc[self.current_step]
        prev = self.df.loc[self.current_step - 1]
        pct_change = float((row['close'] - prev['close']) / (prev['close'] + 1e-9))
        ma5 = float(self.df['close'].rolling(5).mean().iloc[self.current_step]) if self.current_step >= 4 else float(row['close'])
        ma20 = float(self.df['close'].rolling(20).mean().iloc[self.current_step]) if self.current_step >= 19 else float(row['close'])
        ma5_diff = (row['close'] - ma5) / (ma5 + 1e-9)
        ma20_diff = (row['close'] - ma20) / (ma20 + 1e-9)
        news = float(row.get('news_signal', 0.0))
        obs = np.array([pct_change, ma5_diff, ma20_diff, news], dtype=np.float32)
        return obs

    def reset(self):
        self.current_step = 1
        return self._get_obs()

    def step(self, action):
        pos = 0.0
        if action == 1:
            pos = 1.0
        elif action == 2:
            pos = -1.0
        prev_price = float(self.df.loc[self.current_step - 1, 'close'])
        cur_price = float(self.df.loc[self.current_step, 'close'])
        pct_change = (cur_price - prev_price) / (prev_price + 1e-9)
        reward = pos * pct_change
        self.current_step += 1
        done = self.current_step >= len(self.df)
        obs = self._get_obs() if not done else np.zeros(self.observation_space.shape, dtype=np.float32)
        info = {}
        return obs, float(reward), bool(done), info

    def render(self, mode='human'):
        pass
```

```python
# Cell 15 - Smoke-test: pick a ticker with >=30 days, else fall back to first ticker
import numpy as np

tickers = price_df['ticker'].unique()
selected = None
for t in tickers:
    if len(price_df[price_df['ticker'] == t]) >= 30:
        selected = t
        break
if selected is None:
    # if none have 30 days, try to use the first ticker and warn
    selected = tickers[0]
    print("Warning: no ticker has >=30 days; using", selected)

print("Testing env for", selected)
env = TradingEnv(price_df, ticker=selected)
obs = env.reset()
print("obs shape:", obs.shape)
obs, rew, done, info = env.step(1)
print("After one step -> reward:", rew, "done:", done)
```

```python
# Cell 16 - Train a PPO agent (demo). Increase total_timesteps for real runs.
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv
import os

TICKER = selected  # from Cell 15
vec_env = DummyVecEnv([lambda: TradingEnv(price_df, ticker=TICKER)])
model = PPO('MlpPolicy', vec_env, verbose=1, seed=SEED)
model_path = os.path.join(BASE_DIR, 'ppo_trading_model')
print("Training PPO (demo): total_timesteps=20000 (reduce/increase as needed)...")
model.learn(total_timesteps=20000)
model.save(model_path)
print("Saved PPO model to", model_path + ".zip")
```

```python
# Cell 17 - Evaluate the trained policy and plot cumulative reward
import matplotlib.pyplot as plt
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv
import os, numpy as np

model = PPO.load(os.path.join(BASE_DIR, 'ppo_trading_model'), env=DummyVecEnv([lambda: TradingEnv(price_df, ticker=TICKER)]))
env_eval = DummyVecEnv([lambda: TradingEnv(price_df, ticker=TICKER)])
obs = env_eval.reset()
cum_rewards = []
while True:
    action, _ = model.predict(obs, deterministic=True)
    obs, reward, done, info = env_eval.step(action)
    cum_rewards.append(float(reward))
    if done:
        break

cum_rewards = np.array(cum_rewards)
plt.plot(np.cumsum(cum_rewards))
plt.title(f'Cumulative reward ({TICKER})')
plt.xlabel('step')
plt.ylabel('cumulative reward')
plt.grid(True)
plt.show()

print("Total cumulative reward:", float(cum_rewards.sum()))
```
